Consider all integer combinations of $a^b$ for $2 \leq a \leq 5$ and $2 \leq b \leq 5$:
$$
\begin{align}
&2^2 = 4, \quad 2^3 = 8, \quad 2^4 = 16, \quad 2^5 = 32 \\
&3^2 = 9, \quad 3^3 = 27, \quad 3^4 = 81, \quad 3^5 = 243 \\
&4^2 = 16, \quad 4^3 = 64, \quad 4^4 = 256, \quad 4^5 = 1024 \\
&5^2 = 25, \quad 5^3 = 125, \quad 5^4 = 625, \quad 5^5 = 3125
\end{align}
$$
If they are then placed in numerical order, with any repeats removed, we get the following sequence of $15$ distinct terms:
$$
\begin{align}
4, 8, 9, 16, 25, 27 ,32, 64, 81, 125, 243, 256, 625, 1024, 3125.
\end{align}
$$
How many distinct items are in the sequence generated by $a^b$ for $2 \leq a \leq 100$ and $2 \leq b \leq 100$?

```python
def number_combinations(a, b):
    number_lst = []
    for base in range(2, a + 1):
        for exponent in range(2, b + 1):
            number_lst.append(base ** exponent)
    number_lst.sort()
    return len(set(number_lst))
```

```python
number_combinations(100, 100)
```
