# Pentagon Numbers
Pentagonal numbers are generated by the formula, $P_n = n(3n - 1) / 2$. The first ten pentagonal numbers are:
$$
\begin{align}
1, 5, 12, 22, 35, 51, 70 , 92 ,117, 145, \ldots
\end{align}
$$
It can be seen that $P_4 + P_7 = 22 + 70 = 92 = P_8$. However, their difference, $70 - 22 = 48$, is not pentagonal. </br>

Find the pair of pentagonal numbers, $P_j$ and $P_k$, for which their sum and difference are pentagonal and $D = |P_k - P_j|$ is minimised; what is the value of $D$?

```python
import numpy as np
```

```python
def is_pentagonal(num):
    n = (1 + (1 + 24 * num) ** 0.5) / 6
    return n.is_integer()
```

```python
def pentagonal(n):
    return int(n * (3 * n - 1) / 2)
```

```python
pentagonal_lst = [pentagonal(x) for x in range(1, 10 ** 6)]
```

```python
N = 10**4
d = np.inf

for i in range(1, N):
    for j in range(i + 1, N):
        pentagonal_sum = pentagonal(i) + pentagonal(j)
        pentagonal_minus = pentagonal(j) - pentagonal(i)
        
        if is_pentagonal(pentagonal_sum) and is_pentagonal(pentagonal_minus):
            print("O")
            d = min(d, pentagonal_minus)
```
