# Triangular, Pentagonal, and Hexagonal
Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:
$$
\begin{eqnarray}
&\text{Triangle} & T_n = n(n+1)/2 \\
&\text{Pentagonal} & P_n = n(3n-1)/2\\
&\text{Hexagonal} & H_n = n (2n-1)
\end{eqnarray}
$$
It can be verified that $T_{285} = P_{165} = H_{143} = 40755$. </br>

Find the next triangle number that is also pentagonal and hexagonal.

```python
def triangle(n):
    return int(n * (n + 1) / 2)
```

```python
def is_pentagonal(num):
    n = (1 + (1 + 24 * num)**0.5) / 6
    return n.is_integer()
```

```python
def is_hexagonal(num):
    n = (1 + (1 + 8 * num)**0.5) / 4
    return n.is_integer()
```

```python
N = 286

while 1:
    test = triangle(N)
    if is_pentagonal(test) and is_hexagonal(test):
        print(N, triangle(N))
        break
    
    N += 1
```
