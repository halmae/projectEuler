The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be $1 + 2 + 3 + 4 + 5 + 6 + 7 = 28$. The first ten terms would be:
$$
1, 3, 6, 10, 15, 21, 28, 36, 45, 55, \cdots
$$
Let us list the factors of the first seven triangle numbers:
$$
\begin{align}
&\phantom{0}1:\, 1 \\
&\phantom{0}3:\, 1,3 \\
&\phantom{0}6:\, 1,2,3,6 \\
&10:\, 1, 2, 5, 10 \\
&15:\, 1, 3, 5, 15 \\
&21:\, 1, 3, 7, 21 \\
&28:\, 1, 2, 4, 7, 14, 28
\end{align}
$$
We can see that $28$ is the first triangle number to have over five divisors. </br>
What is the value of the first triangle number to have over five hundred divisors?

```python
import numpy as np
from numba import jit, int64
```

```python
def tri_num(N):
    return int(N * (N + 1) / 2)

@jit(int64(int64))
def find_divisors(N):
    test = []
    for num in range(1, tri_num(N) // 2):
        if tri_num(N) % num == 0:
            test.append(num)
            test.append(int(tri_num(N) / num))
    return len(set(test))
```

```python
N = 1000
while find_divisors(N) < 500:
    N += 1
    print(find_divisors(N))
```

```python
from sympy import factorint
from math import prod

def count_divisors(n):
    factorization = factorint(n)
    # Number of divisors is the product of the powers of prime factors plus 1
    return 1 if n == 1 else prod(exponent + 1 for exponent in factorization.values())

def find_triangle_with_divisors(target_divisors):
    n = 1
    while True:
        triangle_number = n * (n + 1) // 2
        divisors = count_divisors(triangle_number)
        if divisors > target_divisors:
            return triangle_number
        n += 1

result = find_triangle_with_divisors(500)
print(result)
```
